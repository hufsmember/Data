```python
-- 1. Allergy Table
CREATE TABLE Allergy (
    allergy_id INT PRIMARY KEY,
    allergy_name VARCHAR(255)
);

-- 2. Food Table
CREATE TABLE Food (
    food_id INT PRIMARY KEY,
    food_name VARCHAR(255),
    storage_method VARCHAR(50),
    allergy_id INT,
    is_vegan BOOLEAN,
    FOREIGN KEY (allergy_id) REFERENCES allergy_table(allergy_id)
);

-- 3. Fridge Contents Table
CREATE TABLE Fridge_contents (
    user_id INT,
    food_id INT,
    quantity INT,
    purchase_date DATE,
    expiry_date DATE,
    PRIMARY KEY (user_id, food_id),
    FOREIGN KEY (food_id) REFERENCES food_table(food_id)
);

-- 4. Product Table
CREATE TABLE Product (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(255),
    price DECIMAL(10, 2),
    category VARCHAR(100),
    image_url TEXT,
    description TEXT
);

-- 5. Recipe Ingredients Table
CREATE TABLE Recipe_ingredients (
    recipe_ingredients_id INT PRIMARY KEY,
    recipe_id INT,
    food_id INT,
    quantity INT,
    FOREIGN KEY (recipe_id) REFERENCES recipes_table(recipe_id),
    FOREIGN KEY (food_id) REFERENCES food_table(food_id)
);

-- 6. Recipes Table
CREATE TABLE Recipes (
    recipe_id INT PRIMARY KEY,
    recipe_name VARCHAR(255),
    cuisine_type VARCHAR(50),
    description TEXT,
    image_url TEXT,
    is_vegan BOOLEAN
);

-- 7. Shopping Lists Table
CREATE TABLE Shopping_lists (
    shopping_list_id INT PRIMARY KEY,
    user_id INT,
    food_id INT,
    quantity INT,
    FOREIGN KEY (user_id) REFERENCES user_data_table(user_id),
    FOREIGN KEY (food_id) REFERENCES food_table(food_id)
);

-- 8. User Allergies Table
CREATE TABLE User_allergies (
    user_id INT,
    allergy_id INT,
    PRIMARY KEY (user_id, allergy_id),
    FOREIGN KEY (user_id) REFERENCES user_data_table(user_id),
    FOREIGN KEY (allergy_id) REFERENCES allergy_table(allergy_id)
);

-- 9. User Data Table
CREATE TABLE User (
    user_id INT PRIMARY KEY,
    user_name VARCHAR(255),
    email VARCHAR(255),
    password VARCHAR(255),
    is_family_exist BOOLEAN,
    gender VARCHAR(10),
    age_group VARCHAR(20),
    is_vegan BOOLEAN,
    preferred_cuisine VARCHAR(50),
    non_preferred_cuisine VARCHAR(50)
    preferred_ingredient VARCHAR(50),
    non_preferred_ingredient VARCHAR(50)
);

-- 10. User Preferences Table
CREATE TABLE user_preferences (
    user_preference_id INT PRIMARY KEY,
    user_id INT,
    recipe_id INT,
    preference_level INT,
    liked BOOLEAN,
    FOREIGN KEY (user_id) REFERENCES user_data_table(user_id),
    FOREIGN KEY (recipe_id) REFERENCES recipes_table(recipe_id)
);
```
