```python
import pandas as pd

# 파일 경로
file_path = r"C:\Users\USER\OneDrive - 한국외국어대학교\바탕 화면\최종 mysql데이터\allergy_table.csv"

# CSV 파일을 읽어 DataFrame 생성
allergy_df = pd.read_csv(file_path)

# DataFrame 출력
print(allergy_df)

# 데이터의 총 개수 출력
print("Total rows:", allergy_df.shape[0])

# 컬럼 구성 출력
print("Columns:", allergy_df.columns.tolist())

# 각 컬럼의 고유한 값들 출력
for column in allergy_df.columns:
    print(f"Unique values in {column}:")
    print(allergy_df[column].unique())
```
